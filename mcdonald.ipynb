# 3. Correlation Matrix

```python
#Credits https://www.kaggle.com/code/prena0808/exploratory-data-analysis-for-nutrition-value

corr = df.drop(['menu_category', 'menu_items'], axis=1).corr() 
plt.figure(figsize=(8, 8))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, 
            linewidths=0.1,annot=True, 
            annot_kws={"size": 8}, square=True)
```

# 5.Unhealthy Menu Options-measured in terms of total fat content, total sugar, cholesterol, or sodium

##### Source : https://www.healthline.com/nutrition/sodium-per-day

#### Checking if any dish has sodium content greater than or equal to 2300mg

```python
df[df['sodium_(mg)']>=2300][['menu_category', 'menu_items', 'per_serve_size', 'energy_(kcal)',
       'protein_(g)', 'total_fat_(g)',
       'cholesterols_(mg)', 'total_carbohydrate_(g)', 'total_sugars_(g)', 'sodium_(mg)']]
```

#### Assuming a person consumes one meal from McDonald then sodium content of more than 1725mg which is 3 fourth of the limit per day could be dangerous.
#### Checking for food items with sodium content more than 1725mg

```python
df[df['sodium_(mg)'].apply(lambda x: 1725 <= x < 2300)][['menu_category', 'menu_items', 'per_serve_size', 'energy_(kcal)',
       'protein_(g)', 'total_fat_(g)',
       'cholesterols_(mg)', 'total_carbohydrate_(g)', 'total_sugars_(g)', 'sodium_(mg)']]
```

```python
# Create the boxplot
plt.figure(figsize = (12, 6))

ax = sns.boxplot(x='menu_category', y='sodium_(mg)',data=df,flierprops={'markerfacecolor': 'red', 'marker': 'o'})

# Rotate x-axis labels
plt.xticks(rotation=45)

# Set title
plt.title('Total Sodium Mark')

# Set background color
fig = plt.gcf()
fig.set_facecolor('black')
ax.set_facecolor('black')

# Set tick colors
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Set label colors
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')

# Set grid color
#ax.yaxis.grid(color='white')

# Display the plot
plt.show()
```

#### Factors that can contribute to the gap between the estimated value and the true value of 402.05 kcal include:

#### 1.Measurement accuracy: There can be slight variations in the accuracy of measuring the exact amounts of macronutrients in a food item, leading to small discrepancies in the calculated energy content.

#### 2.Food processing and cooking methods: Food processing techniques and cooking methods can affect the availability and digestibility of nutrients, which can impact the actual energy content of the dish.

#### 3.Nutrient interactions: The presence of certain nutrients or food components in combination can affect the absorption and utilization of energy from the macronutrients, potentially altering the overall energy content.

#### 4.Labeling inaccuracies: Nutritional information provided on food labels or in nutritional databases might not always be completely accurate, as there can be errors in data entry or variations in the composition of different batches or brands of the same food.
